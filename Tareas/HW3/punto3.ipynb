##Primer punto
Función para calcular el seno:

````python
def numsin(n, x):
    term_sum=[0]*(n+1) # Guarda los terminos de la suma
    term_sum[1]=1.0*x #n=1
    suma= term_sum[1] # empieza la suma con el primer termino 
    for i in range (2,n+1):
        term_sum[i]= ((-x**2)/((2*i-1)*(2*i-2)))*(term_sum[i-1])
        suma+= term_sum[i]
    return suma# Guarda los terminos de la suma
````

A continuación se encuentra el segundo punto corriendo hasta el termino 79 del linspace.

````python
a= linspace(0.,45,100)
iteraciones=[]
errores_abs=[]
a_usados=[]
sumas_x=[]
reales =[]
for i in range(80): #len(a)
    real=np.sin(a[i]) # valor real del sin(x)
    m=1 # nuemro de terminos de la suma 
    nuevo= numsin(m,a[i]) # El primer nuevo a comparar
    valabs=abs(real- nuevo)
   
    while valabs > 0.01:
        m=m+1
        nuevo= numsin(m,a[i])
        valabs=abs(real- nuevo) 

    iteraciones += [m] # Guarda el numero de terminos de la suma que cumplen la condición
    errores_abs += [valabs] # Guarda últimos valores de diferencia absoluta respecto al real
    a_usados += [a[i]] # guarda el valor de x
    sumas_x += [nuevo] # guarda el resultado de la suma para ese x
    reales += [real] # guarda el sin(x)
    print i, a_usados[i], m, valabs, "SUMA", nuevo
````

Las gráficas para entender porque el metodo corre hasta 79 con 0.01 como ladiferencia mínima se encuentran a continuación. Las primeras van solamente hasta

````python
plt.figure(figsize=(13,10))

plt.subplot(311)
plt.title(' Resultados para los valores del linspace hata el [79]')
plt.xlabel('Valores de x para sin(x)')
plt.ylabel('Diferencia del valor real al de la sumatoria')
plt.plot(a_usados,errores_abs)

plt.subplot(312)
plt.xlabel('Valores de x para sin(x)')
plt.ylabel('Diferencia del valor real al de la sumatoria')
plt.plot(a_usados, iteraciones)

plt.subplot(313)
plt.xlabel('Valores de x para sin(x)')
plt.ylabel('Valor de la sumatoria')
plt.plot(a_usados, sumas_x, c='m')#, plt.plot(a_usados, reales, c='g')
````
